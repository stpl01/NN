```python
def neuron(x, w, phi):
	if len(x) != len(w):
		raise ArithmeticError("x and w have to have the same length!")
	
	fkt = lambda sigma: 0 if sigma < phi else 1
	
	sigma = 0
	for i in range(len(x)):
		sigma += x[i]*w[i]
		
	return fkt(sigma)


print(neuron([1, 1], [0.5, -1], -1))


def neuron_complicated(x, w, phi):
    if len(x) != len(w):
        raise ArithmeticError("x and w have to have the same length!")
    
    return 0 if sum(xn*wn for xn, wn in zip(x, w)) < phi else 1
```

```python
#importiert zwei tolle Module
import numpy as np
import matplotlib.pyplot as plt

#Funktiongleichung für die Trennungsgerade
def dfkt ( Sigma, Steigung):
    ultima = Sigma * Steigung
    return ultima

#Variabelinput
x1 = 0.9
x2 = 1
x3 = 0.76
x4 = 0.7
x5 = 0.74
x6 = 0.78
x7 = 0.82

y1 = 0.25
y2 = 0.25
y3 = 0.4
y4 = 0.1
y5 = 0.4
y6 = 0.5
y7 = 0.5

"""
Aus zwei Werten von jeweils einer Person wird ein Wert ermittelt,
mit welchem man besser die Werte von den anderen vergleichen kann.
"""
xy1 = y1 / x1
xy2 = y2 / x2
xy3 = y3 / x3

xy4 = y4 / x4
xy5 = y5 / x5
xy6 = y6 / x6
xy7 = y7 / x7

#mliste ist ein array mit einer Liste von den erstellten Werten aus dem Input
mliste = [xy4, xy5, xy6, xy7]
MannDurchschnitt = xy4
for i in mliste:
    #Der kleinste Wert wird ermittelt und zur Kontrolle werden die Werte dann ausgegeben
    print(MannDurchschnitt)
    if i < MannDurchschnitt:
        MannDurchschnitt = i
#wliste ist ein array mit einer Liste von den erstellten Werten aus dem Input
wliste = [xy1, xy2, xy3]
FrauDurchschnitt = xy1
for i in wliste:
    #Der groesste Wert wird ermittelt und zur Kontrolle werden die Werte dann ausgegeben
    print(FrauDurchschnitt)
    if i > FrauDurchschnitt:
        FrauDurchschnitt = i
        
#Nun haben wir die Steigung ermittelt,
#indem wir den Wert zwischen FrauDurchschnitt und MannDurchschnitt nehmen
Steigung = (FrauDurchschnitt - MannDurchschnitt)/2 + MannDurchschnitt

    
    
#Zur Kontrolle wird der Wert ausgegeben
print(Steigung)

#Nun werden die ursprünglichen Input Werte genommen in einem array
FX = [x1,x2,x3]
MX = [x4,x5,x6,x7]
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        
FY = [y1,y2,y3]
MY = [y4,y5,y6,y7]

#x-Achse wird erstellt
X = np.linspace(0, 2, 50, endpoint=True)

#Trennungsgerade wird ausgerufen
G = [dfkt(i, Steigung) for i in X]
'''
arrays werden aufgerufen um Punkte zu erstellen, welche farblich gekennzeichnet sind
blau -> Mann
lila -> Frau
'''
plt.scatter(FX,FY,label="fkt",color = "purple")
plt.scatter(MX,MY,label="wow",color = "blue")
#Trennungsgerade wird in die Funktion gepackt
plt.plot(X, G)
#Funktion wird dargestellt
plt.show()
```

```python
# Module werden importiert
# -> Matplotlib -> pyplot als plt
# -> Matplotlib -> Slider
# -> Matplotlib -> Button
# -> numpy als np
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.widgets import Slider, Button

# Um Schieberegler anwenden zu können, muss das Bild extern gezeichnet werden.
# Dazu dient der folgende Befehl.
%matplotlib qt

# Gewichte der Männer werden festgelegt
MannGewicht0 = 0.35
MannGewicht1 = 0.4
MannGewicht2 = 0.5
MannGewicht3 = 0.5

# Größen der Männer werden festgelegt
MannGroesse0 = 0.7
MannGroesse1 = 0.74
MannGroesse2 = 0.78
MannGroesse3 = 0.82

# Gewichte der Frauen werden festgelegt
FrauGewicht0 = 0.25
FrauGewicht1 = 0.25
FrauGewicht2 = 0.35

# Größen der Frauen werden festgelegt
FrauGroesse0 = 0.66
FrauGroesse1 = 0.7
FrauGroesse2 = 0.76

# Intialparameter bzw. Y-Achsenabschnitt und Steigung werden festgelegt
w1 = 0.825
w2 = -0.25

# Frauenwerte werden in einem Array gespeichert
FX = [FrauGroesse0, FrauGroesse1, FrauGroesse2]
FY = [FrauGewicht0, FrauGewicht1, FrauGewicht2]

# Männerwerte werden in einem Array gespeichert
MX = [MannGroesse0, MannGroesse1, MannGroesse2, MannGroesse3]
MY = [MannGewicht0, MannGewicht1, MannGewicht2, MannGewicht3]

# Lineare Funktion wird beschrieben: mx + b
def function(x, w1, w2):
    return w1 * x + w2

# X-Werte werden für die Funktion erstellt
x = np.linspace(0, 1, 1000)


# Erstellt die Figur und den Graphen
fig, ax = plt.subplots()
line, = plt.plot(x, function(x, w1, w2), lw=2)
ax.set_xlabel('x')
ax.set_ylabel('y')

# Verschiebt das Koordinatensystem, um für den Regler Platz zu machen
plt.subplots_adjust(left=0.25, bottom=0.25)

# Horizontaler Schieberegler für Y-Achsenabschnitt wird erstellt.
# Größe des Reglers (left, bottom, width, height)
plt.scatter(FX, FY, label="fkt", color="purple")
plt.scatter(MX, MY, label="wow", color="blue")
axYAchsenAbschnitt = plt.axes([0.25, 0.1, 0.65, 0.03])
yAchsenabschnittRegler = Slider(
    ax=axYAchsenAbschnitt,
    label='Y-Achsenabschnitt',      # definiert Namen des Reglers
    valmin=-1.5,                    # Minimum-Wert des Y-Achsenabschnitts-Reglers
    valmax=1.5,                     # Maximumwert des Y-Achsenabschnitts-Reglers
    valinit=w2,                     # Initialisierungswert
)

# Vertikaler Schieberegler für Steigung wird erstellt.
# Größe des Reglers (left, bottom, width, height)
axSteigung = plt.axes([0.1, 0.25, 0.0225, 0.63])
steigungsRegler = Slider(
    ax=axSteigung,
    label="Steigung",               # definiert Namen des Reglers
    valmin=-5,                       # Minimum-Wert des Steigung-Reglers
    valmax=5,                      # Maximumwert des Steigung-Reglers
    valinit=w1,                     # Initialisierungswert
    orientation="vertical"          # Andere Ausrichtung des Reglers
)

# Updatet die Schieberegler-Werte, wenn Schieberegler geändert
def update(val):
    line.set_ydata(function(x, steigungsRegler.val, yAchsenabschnittRegler.val))    # Die Werte der beiden Regler werden entnommen und als neue Werte gesetzt
    fig.canvas.draw_idle()


# Registriert Veränderung im Schieberegler und ruft die Update-Funktion auf
yAchsenabschnittRegler.on_changed(update)
steigungsRegler.on_changed(update)

# Wenn die Veränderung unerwünscht war, kann man hier den Wert zurücksetzen auf die ursprünglich gegebenen Werten
axReset = plt.axes([0.8, 0.025, 0.1, 0.04])
button = Button(axReset, 'Reset', hovercolor='0.975') # Knopfeigenschaften : Name und Farbe beim drüberhovern

# Setzt die Werte zurück
def reset(event):
    yAchsenabschnittRegler.reset()
    steigungsRegler.reset()
# Registriert Knopfdruck
button.on_clicked(reset)
plt.show()
```
